### Q1: â€˜hpâ€™ and â€˜mpgâ€™.â€˜hpâ€™ and â€˜time-to-60â€™.â€˜hpâ€™ and â€˜weightlbsâ€™.â€˜hpâ€™ and â€˜cylindersâ€™. scatter plotdersâ€™.

##### will use matplotlip plt

```python
plt.scatter(df['hp'],df['mpg'])
plt.xlabel('horsepower')
plt.ylabel('miles_per_galons')
plt.title('hp vs mpg')
plt.show()
#this is just an example how we can plot scatter plot for single plot, now we will use set for all plots together
```

```python
fig,axs = plt.subplots(2,2)

axs[0,0].scatter(df['hp'],df['mpg'])
axs[0,0].set_title('horsepower vs miles_per_galons')

axs[0,1].scatter(df['hp'],df['time-to-60'])
axs[0,1].set_title('horsepower vs time_in_secs')

axs[1,0].scatter(df['hp'],df['weightlbs'])
axs[1,0].set_title('horsepower vs weightlbs')

axs[1,1].scatter(df['hp'],df['cylinders'])
axs[1,1].set_title('horsepower vs cylinders')

plt.subplots_adjust(wspace=1,hspace=1)
```

##### now, we'll use seaborn to plot all together

```python
fig,axs = plt.subplots(2,2, figsize=(15,8))

sns.scatterplot(data=df,x='hp',y='mpg',ax=axs[0,0])
axs[0,0].set_xlabel('horsepower',fontsize=11)
axs[0,0].set_ylabel('miles_per_galons',fontsize=11)

sns.scatterplot(data=df,x='hp',y='time-to-60',ax=axs[0,1])
axs[0,1].set_xlabel('horsepower',fontsize=11)
axs[0,1].set_ylabel('time_in_secs',fontsize=11)

sns.scatterplot(data=df,x='hp',y='weightlbs',ax=axs[1,0])
axs[1,0].set_xlabel('horsepower',fontsize=11)
axs[1,0].set_ylabel('weight_in_lbs',fontsize=11)

sns.scatterplot(data=df,x='hp',y='cylinders',ax=axs[1,1])
axs[1,1].set_xlabel('horsepower',fontsize=11)
axs[1,1].set_ylabel('cylinders',fontsize=11)

plt.tight_layout() #function in Python is used to automatically adjust the subplot parameters for a 
                    #Matplotlib figure to ensure that the plot elements do not overlap
#plt.show()
```

### Q2: Create scatter plots with between (mpg-hp, mpg-weight), and (hp-mpg, hp-time-to-60) with respect to each brand

##### Q3: Create bar plots for the following, make sure all the plots appear in the same plane with x-labels and y-labels evenly spaced.

### Q4: Create a bar plot that shows the visual representation of hp, mpg, weight, time-to-60 with respect to the number of cylinders for each of the brands.

Create a bar plot that shows the visual representation of hp, mpg, weight and time-to-60 with respect to the years, for each brand.

##### using matplotlib 

```python
fig,axs = plt.subplots(2,4)

axs[0,0].bar(df['brand'],df['cylinders'])
axs[0,0].bar(df['brand'],df['hp'])
axs[0,0].set_title(' cylinders vs horsepower')
plt.xticks(rotation=90)


axs[0,1].bar(df['brand'],df['cylinders'])
axs[0,1].bar(df['brand'],df['mpg'])
axs[0,1].set_title(' cylinders vs miles/galon')
plt.xticks(rotation=90)


axs[0,2].bar(df['brand'],df['cylinders'])
axs[0,2].bar(df['brand'],df['weightlbs'])
axs[0,2].set_title(' cylinders vs weight')
plt.xticks(rotation=90)


axs[0,3].bar(df['brand'],df['cylinders'])
axs[0,3].bar(df['brand'],df['time-to-60'])
axs[0,3].set_title(' cylinders vs time')
plt.xticks(rotation=90)


axs[1,0].bar(df['brand'],df['year'])
axs[1,0].bar(df['brand'],df['hp'])
axs[1,0].set_title('year vs horsepower')
plt.xticks(rotation=90)


axs[1,1].bar(df['brand'],df['year'])
axs[1,1].bar(df['brand'],df['mpg'])
axs[1,1].set_title('year vs miles/galon')
plt.xticks(rotation=90)


axs[1,2].bar(df['brand'],df['year'])
axs[1,2].bar(df['brand'],df['weightlbs'])
axs[1,2].set_title('year vs weight')
plt.xticks(rotation=90)


axs[1,3].bar(df['brand'],df['year'])
axs[1,3].bar(df['brand'],df['time-to-60'])
axs[1,3].set_title('year vs time')
plt.xticks(rotation=90)


plt.tight_layout(pad=1,h_pad=1,w_pad=1)
plt.subplots_adjust(wspace=1,hspace=1)
```

##### using seaborn 

```python
fig,axs = plt.subplots(2,4, figsize=(15,8))

sns.barplot(data=df,x='cylinders',y='hp',ax=axs[0,0])
axs[0,0].set_xlabel('cylinders',fontsize=11)
axs[0,0].set_ylabel('hp',fontsize=11)

sns.barplot(data=df,x='cylinders',y='mpg',ax=axs[0,1])
axs[0,1].set_xlabel('cylinders',fontsize=11)
axs[0,1].set_ylabel('mpg',fontsize=11)

sns.barplot(data=df,x='cylinders',y='weightlbs',ax=axs[0,2])
axs[0,2].set_xlabel('cylinders',fontsize=11)
axs[0,2].set_ylabel('weight',fontsize=11)

sns.barplot(data=df,x='cylinders',y='time-to-60',ax=axs[0,3])
axs[0,3].set_xlabel('cylinders',fontsize=11)
axs[0,3].set_ylabel('time',fontsize=11)

sns.barplot(data=df,x='year',y='hp',ax=axs[1,0])
axs[1,0].set_xlabel('year',fontsize=11)
axs[1,0].set_ylabel('hp',fontsize=11)

sns.barplot(data=df,x='year',y='mpg',ax=axs[1,1])
axs[1,1].set_xlabel('year',fontsize=11)
axs[1,1].set_ylabel('mpg',fontsize=11)

sns.barplot(data=df,x='year',y='weightlbs',ax=axs[1,2])
axs[1,2].set_xlabel('year',fontsize=11)
axs[1,2].set_ylabel('weight',fontsize=11)

sns.barplot(data=df,x='year',y='time-to-60',ax=axs[1,3])
axs[1,3].set_xlabel('year',fontsize=11)
axs[1,3].set_ylabel('time',fontsize=11)

plt.tight_layout() #function in Python is used to automatically adjust the subplot parameters for a 
                    #Matplotlib figure to ensure that the plot elements do not overlap
#plt.show()
plt.xticks(rotation=90)
```
